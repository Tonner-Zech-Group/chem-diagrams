```python
from PIL import Image


def add_whitespace_to_left(image, n_pixels):

    img = image
    width, height = img.size

    # Create a new image with extra whitespace on the left
    new_width = width + n_pixels  # Add n_pixels of whitespace
    new_img = Image.new("RGBA", (new_width, height), (255, 255, 255, 0))  # Transparent background
    new_img.paste(img, (n_pixels, 0))  # Paste original image with an offset of 20 pixels

    return new_img

# Base directory
base_path = os.path.join("..", "docs", "img")

# Image filenames
filenames = [
    "example_styles_I.png",
    "example_styles_II.png",
    "example_styles_III.png",
    "example_styles_IIII.png",
    "example_styles_V.png",
]

# Load images
images = [Image.open(os.path.join(base_path, name)) for name in filenames]

images[4] = add_whitespace_to_left(images[4], (images[0].size[0] - images[4].size[0]))

# Ensure all images are the same size (resize to first image size if needed)
width, height = images[0].size
images = [img.resize((width, height)) for img in images]

# Create new blank image (2x3 grid)
merged_image = Image.new("RGBA", (width * 2, height * 3))

# Paste images into grid
merged_image.paste(images[0], (0, 0))              # Top-left
merged_image.paste(images[1], (width, 0))          # Top-right
merged_image.paste(images[2], (0, height))         # Mid-left
merged_image.paste(images[3], (width, height))     # Mid-right
merged_image.paste(images[4], (0, 2*height))   # Bottom-left

# Save output
output_path = os.path.join(base_path, "example_styles.png")
merged_image.save(output_path)

print(f"Merged image saved to: {output_path}")
```

```python
dia = EnergyDiagram(style="twosided")
dia.draw_path(x_data=[0, 1, 2], y_data=[10, 55, 0], color="blue", path_name="Path A")
dia.draw_path(x_data=[2, 3, 4], y_data=[0, 50, -5], color="red",  path_name="Path B")

# Both paths share y=0 at x=2
dia.merge_plateaus(
    x=2,                        # x-position of the shared plateau in data coordinates
    path_name_left="Path A",    # name of the left path to merge (must match the path_name used in draw_path)
    path_name_right="Path B",   # name of the right path to merge (must match the path_name used in draw_path)
    gap_scale=1.0,              # width of the gap between the two half-bars
    stopper_scale=1.0,          # size of the diagonal tick marks
    angle=30,                   # angle of the tick marks in degrees
)

dia.add_numbers_auto()
dia.set_xlabels(["P1", "TS1", "E", "TS2", "P2"])
dia.ax.set_ylabel("Energy / kJ mol$^{-1}$", fontsize=8)

dia.fig.savefig(os.path.join("..","docs","img","example_merge_plateaus.png"),format="png", bbox_inches="tight")
dia.show()
```
