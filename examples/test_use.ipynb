```python
dia1 = EnergyDiagram(
    figsize=(6,3),
    extra_y_margin=(-0.1,0.15), # extra space above and below the highest and lowest points (relative to the total height)
    extra_x_margin=(0,0.5), # extra space to the left and right of the first and last points (absolute units)
    fontsize=8, 
    verbose=True, # doesnt do much right now
)

dia1.draw_path(
    x_data=[0,1,2,3,4,5],
    y_data=[0,-5.3,2,-3,5,0],
    color='blue',
    linetypes=[1,1,1,1,0], # 1 = dashes, 0 = dashes with gap
    path_name="Reaction 1",
    show_numbers=True # can be turned off to just store the path without numbering it when using a numbering method
)

dia1.draw_path(
    x_data=[0,1,3,4,5],
    y_data=[0,-7,-2,5.5,0],
    color='red',
    linetypes=[1,0,1,0],
    path_name="Reaction 2",
    show_numbers=True 
)

dia1.draw_path(
    x_data=[0,1,2,3,4,5],
    y_data=[-1.5,-6,3.5,-2,5.5,0],
    color='orange',
    linetypes=[1,0,1,0,1],
    path_name="Reaction 3",
    show_numbers=True 
)

###green
dia1.draw_path(
    x_data=[0,1,2,3,4,5],
    y_data=[1.5,-4.5,3.0,-2.5,5.5,0],
    color='green',
    linetypes=[1,1,1,1,0],
    show_numbers=False
)


dia1.set_xlabels(["A", "B", "C", "D", "F\n+H"],labelplaces=[0,1,2,3,5])

dia1.draw_difference_bar(
    5.5, # x position
    [-5, 0], # y positions
    r"$\Delta E$: ", 
    left_side=False, # can be put on left or right side
)

dia1.draw_difference_bar(
    3.5, # x position
    [-7, -3], # y positions
    r"$\Delta I$: ", 
    left_side=True, # can be put on left or right side
    color="purple",
    arrowstyle="<->",
    x_whiskers=(1, 3),
)

dia1.add_numbers_auto(
    x_min_max=(0,3)
) 

dia1.add_numbers_average(
    x_min_max=4,
    color="#885398"
)

dia1.legend(
    fontsize=7,
)


dia1.set_diagram_style("halfboxed")
dia1.ax.set_ylabel("Energy (kJ/mol)", fontsize=8)
dia1.show()
```

```python
from chemdiagrams import EnergyDiagram
import os

x = [0,1,2,3,4]
y_data_1 = [0,-14,44,11,-34]
y_data_2 = [0,-16,59,15,-25]


dia = EnergyDiagram(
    style="twosided",
    dpi=150,
    )

dia.add_xaxis_break(1, gap_scale=1, stopper_scale=1, angle=-30)

dia.add_yaxis_break(10)

dia.set_xlabels(["E", "I1", "TS", "I2", "P"])


dia.draw_path(
    x, y_data_1, "blue",
    path_name="Blue path",
    linetypes=[1,1,1,-1],
    
)




dia.draw_path(
    x, y_data_2, "red",
    path_name="Red path",
    linetypes=[1,1,1,-1],

)



dia.draw_difference_bar(
    1.2, (y_data_1[2], y_data_2[2]),
    r"$\Delta E_\mathrm{A}$: ",
    left_side=True,
    x_whiskers=(2,2),
    whiskercolor="red"
)


dia.bars[0].whisker_1.set_color("blue")


dia.legend(fontsize=6)
dia.ax.set_ylabel(r"$\Delta E$ in kj mol$^{-1}$", fontsize=8)


penguin = os.path.join("figures", "penguin.png")
penguin_TS = os.path.join("figures", "penguin_TS.png")
penguin_II = os.path.join("figures", "penguin_2.png")
ester = os.path.join("figures", "ester_1.png")


dia.add_numbers_auto(fontsize=8)

dia.add_image_in_plot(
    penguin,
    (1, 10),
    vertical_alignment="bottom",
    height=10,
    framed=True,
    frame_color="green"
)

dia.add_image_series_in_plot(
    [penguin, penguin_TS, ester],
    img_x_places=[0,2,3],
    y_offsets=[0,0,0],
    img_series_name="1",
    framed=[True, False, False],
    frame_colors=["black", "", ""],
    y_placement="auto",
    proportional_scaling=True,
    width=0.5,
    
)

print(ann := dia.ax_objects.xaxis_breaks["1.0"].stopper_1)





dia.show()

```

```python
from chemdiagrams import EnergyDiagram

dia = EnergyDiagram(style="twosided", dpi=150)

dia.draw_path(
    [0,1,2], [-30,35,0], "blue",
    path_name="Blue path",
    linetypes=[1,1],
)



dia.draw_path(
    [2,3,4], [0, 40, -35], "red",
    path_name="Red path",
    linetypes=[1,1],
)

dia.merge_plateaus(
    2, "Blue path", "Red path",
)

mp = dia.lines["Red path"].plateaus["2.0"]
print(mp)

mp.plateau_left.set_color("green")
mp.plateau_right.set_color("black")
mp.stopper_left.arrow_patch.set_lw(3)

dia.add_numbers_average(color="black")

dia.legend(fontsize=5)

dia.add_xaxis_break(2)
```
