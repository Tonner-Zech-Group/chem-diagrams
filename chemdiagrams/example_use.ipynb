```python
from energy_diagram import EnergyDiagram
```

# Minimal use example

```python
dia1 = EnergyDiagram(
                    figsize=(6,3),
                    )

dia1.draw_path(
    x_data=[0,1,2,3,4,5],
    y_data=[0,5,2,7,3,0],
    color='blue',
)

dia1.add_numbers_naive()

dia1.set_xlabels(["A", "B", "C", "D", "E", "F"])
dia1.ax.set_ylabel("Energy (kJ/mol)")
```

# More options

```python



dia1 = EnergyDiagram(
                    figsize=(6,3),
                    extra_y_margin=(-0.1,0.15), # extra space above and below the highest and lowest points (relative to the total height)
                    extra_x_margin=(0,0.5), # extra space to the left and right of the first and last points (absolute units)
                    fontsize=8, 
                    verbose=True # doesnt do much right now
                    )

dia1.draw_path(
    x_data=[0,1,2,3,4,5],
    y_data=[0,-5,2,-3,5,0],
    color='blue',
    linetypes=[1,0,1,1,0], # 1 = dashes, 0 = dashes with gap
    path_name="Reaction Pathway",
    show_numbers=True # can be turned off to just store the path without numbering it
)

dia1.draw_path(
    x_data=[0,1,3,4,5],
    y_data=[0,-7,-2,5.5,0],
    color='red',
    linetypes=[1,0,1,0],
    path_name="Reaction Pathway 2",
    show_numbers=True 
)
dia1.set_xlabels(["A", "B", "C", "D", "F\n+H"],labelplaces=[0,1,2,3,5])

dia1.draw_difference_bar(
                        5.5, # x position
                        [-5, 0], # y positions
                        r"$\Delta E$: ", 
                        left_side=False # can be put on left or right side
                        )

dia1.add_numbers_stacked(
                        x_min_max=(0,3), # only add numbers for paths within this x range (inclusive)
                        sort_by_energy=True # sort the numbering by energy (lowest to highest)
                        )


dia1.show()
```
